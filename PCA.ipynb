```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
from mpl_toolkits.mplot3d import Axes3D
from numpy import linalg as LA

def plot_hist(x, t):
    plt.hist(x)
    plt.title(t)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.xticks()
    plt.yticks()
    plt.show()

def find_singular(x):
    #center the data
    x_std=x-x.mean()
    
    pca = decomposition.PCA(n_components=x.shape[0])
    pr_comps=pca.fit_transform(x_std)
    print("singular values:\n",np.round(pca.singular_values_,4))
    
    print("Energy contributed by singular values:\n",
          np.round(pca.singular_values_/sum(pca.singular_values_),4)*100)
def PCA(x,k):
    '''
        input: dataframe:x, number of principal components n
        output: transformed data in new dimensions
    '''
    #center the data
    x_std=x-x.mean()
    
    #create pca object
    pca = decomposition.PCA(n_components=k)
    
    #find principal components
    pr_comps=pca.fit_transform(x_std)
    
    #plot the pricipal components
    x1=pr_comps[:,0]
    x2=pr_comps[:,1]
    plt.scatter(x1,x2)
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.title("Plot of the data in 2 dimensions")
    plt.show()
    
    
    #generate a 100000 samples of k values from N(0,1)
    rand_nos=np.random.normal(0,1,200000)
    #reshape to matrix of 2x10000
    Z=pd.DataFrame(rand_nos.reshape(2,100000))
    
    #find the maximum numbers of samples
    Z_max=Z.max()
    print("Zmax:", Z_max.max())
    print("Zmin",Z_max.min())
    
    #plot histogram
    plot_hist(Z_max,"Histogram of maximum of samples")

    #find the correlated random numbers
    Z_corr=pd.DataFrame(pr_comps.dot(Z))
    print(Z_corr.shape)
    
    
    #find Z_corr max
    Z_corr_max=Z_corr.max()
    print("Zcorrmax:",Z_corr_max.max())
    print("Zcorrmin:",Z_corr_max.min())
    
    #plot the histogram
    plot_hist(Z_corr_max,"Histogram of maximum of correlated samples")
    
    
    
        
    
    
    
    
    
    
    
    #print("singular values:",pca.singular_values_)
    #print("principal components",pca.components_)
    
    #PLOT
    
    
        
#convert the text file to dataframes
df = pd.read_table('vt-data.txt',header=None, delimiter=',')
print(df.shape)
#find_singular(df)
PCA(df,2)

```
