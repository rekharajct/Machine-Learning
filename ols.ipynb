```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

def least_square_model(df,m,n):
    '''
    Find the ordinary least squares fit of data using Python  statsmodel on different 
    data sizes.
    Inputs: 
        df:Dataframe with first column X and second column Y
        m: lower index of dataset
        n: upper index of dataset   
        
    '''
    N=df.shape[0] #size of dataset
    X=df['X'][m:n] #X is the first m to n rows of the column X dataframe
    Y=df['Y'][m:n] #Y is the first m to n rows of the column Y dataframe
    
    X_range=X #store the X values for plotting the graph
    
    #Add bias
    X = sm.add_constant(X)
    #fit the model using least squares
    est = sm.OLS(Y, X).fit()
    #print the summary
    print(est.summary())
    
    #find the parameters
    a=est.params[0] 
    b=est.params[1] 
    print("a=",a,"b=",b)
    
    #predict using estimated parameters
    pred = a+b*X['X']
    
    #find residual sum of squares
    RSS=sum((Y-pred)*(Y-pred))
    print("RSS=",RSS)
    
    #plot the graph
    plt.figure(figsize=(10,10))
    plt.scatter(X_range,Y,color='b',label="Data")
    plt.plot(X_range,pred,color='r',linewidth=1,label="a+bt")
    plt.legend()
    plt.xlabel("t")
    plt.ylabel("f(t)")
    plt.show()

'''
The text files contains data in the following format where 1st column is the X value and 
second column is Y

1.0	3.1
2.0	5.1
3.0	7.2
4.0	9.2
5.0	11.2
6.0	13.2
7.0	15.3
8.0	17.3
9.0	19.3

The three text files Data1.txt, Data2.txt, Data3.txt contain 200 data items.
'''
    
#convert the text file to dataframes

df1 = pd.read_table('Data1.txt',delim_whitespace=True, names=('X', 'Y'))
df2 = pd.read_table('Data2.txt',delim_whitespace=True, names=('X', 'Y'))
df3 = pd.read_table('Data3.txt',delim_whitespace=True, names=('X', 'Y'))

#Fit Data1
least_square_model(df1,0,50)
least_square_model(df1,0,100)
least_square_model(df1,0,200)

#Fit Data2

least_square_model(df2,0,50)
least_square_model(df2,0,100)
least_square_model(df2,0,200)

#Fit Data3
least_square_model(df3,0,50)
least_square_model(df3,0,100)
least_square_model(df3,0,200)
```
