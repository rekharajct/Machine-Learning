```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly

def polynomial_regression(M,N):
    #Generate M random numbers 
    #X=np.sort(np.random.uniform(0,0.001, M))
    np.random.seed(0)
    X=np.linspace(0,.001,M)
    print("X")
    print("-----------")
    print(X)
    Y_original=np.sin(X*2000*np.pi)
    Y=np.sin(X*2000*np.pi) + np.random.normal(0,0.3,M)
    print("Y")
    print("-----------")
    print(Y)
    
    coeffs=poly.polyfit(X,Y,N-1)
    print("Coefficients")
    print("-----------")
    print(coeffs)
    
    pred = poly.polyval(X,coeffs)
    print ("Predicted Y:")
    print("-----------")
    print(pred)
    
    #find residual sum of squares
    RSS=sum((Y-pred)*(Y-pred))
    print("RSS=",RSS)
    
    #plot the figure for Y_original
    plt.figure(figsize=(6,6))
    plt.xlim(0,0.001)
    plt.scatter(X,Y_original,color='b',label="Data")
    plt.plot(X,Y_original,color='g',linewidth=1,label="ground truth")
    plt.plot(X,Y,color='r',linewidth=1,label="curve with noise")
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
    #plot the image
    
    plt.figure(figsize=(6,6))
    plt.xlim(0,0.001)
    plt.scatter(np.sort(X),Y,color='b',label="Data")
    
    #plt.plot(np.sort(X),coeffs[0]*np.sort(X)+coeffs[1],color='r',linewidth=1,label="prediction")
    #plt.plot(X,pred,color='r',linestyle="--",linewidth=2,label="prediction")
    plt.plot(X,pred,color='r',linewidth=2,label="prediction")
    
    plt.plot(X,Y,color='g',linewidth=1,label="ground truth")
    
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

polynomial_regression(10,2)
polynomial_regression(10,5)
polynomial_regression(10,10)
```
